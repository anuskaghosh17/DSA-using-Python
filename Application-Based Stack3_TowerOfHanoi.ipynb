```python
# Tower of Hanoi using Stack
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return None

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        return None

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def __str__(self):
        return str(self.items)

def move_disk(source, target, source_name, target_name):
    disk = source.pop()
    target.push(disk)
    print(f"Move disk {disk} from {source_name} to {target_name}")

def move_tower(n, source, target, auxiliary, source_name, target_name, auxiliary_name):
    if n > 0:
        # Move n-1 disks from source to auxiliary
        move_tower(n - 1, source, auxiliary, target, source_name, auxiliary_name, target_name)
        
        # Move the nth disk from source to target
        move_disk(source, target, source_name, target_name)
        
        # Move the n-1 disks from auxiliary to target
        move_tower(n - 1, auxiliary, target, source, auxiliary_name, target_name, source_name)

# Initialize towers
n = int(input("Enter the number of disks: "))
source = Stack()
target = Stack()
auxiliary = Stack()

# Populate the source tower with disks (largest at bottom)
for i in range(n, 0, -1):
    source.push(i)

print("Initial state:")
print(f"Source: {source}")
print(f"Target: {target}")
print(f"Auxiliary: {auxiliary}")

# Solve the Tower of Hanoi
move_tower(n, source, target, auxiliary, "Source", "Target", "Auxiliary")

print("Final state:")
print(f"Source: {source}")
print(f"Target: {target}")
print(f"Auxiliary: {auxiliary}")
```
