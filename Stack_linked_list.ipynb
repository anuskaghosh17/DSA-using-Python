```python
class StackNode:
    def __init__(self,Newdata,link):   #declaration of node structure
        self.data = Newdata
        self.next = link

class Stack:
    # create an empty stack
    def __init__(self):
        self.top = None

    def isempty(self):
        return self.top is None

    def push(self):
        # push and return the top element in the stack
        # node = self.top
        self.top = StackNode(self.data, self.top)

    def pop(self):
        node = self.top
        self.top = self.top.next
        return node.data

    def peek(self):
        return self.top.data

    def display(self,Newdata):
        currentNode = self.top
        print()
        while currentNode is not None:
            print(currentNode.data)
            currentNode = currentNode.next

s = Stack()
while(True):
    print("\n Program to implement stack using Linked list")
    print("1.Push")
    print("2.Pop")
    print("3.Peek")
    print("4.Display")
    print("5.Exit")
    print("_________________________")
    choice = int(input("Enter your choice:"))
    if choice == 1:
        num = int(input("Enter the data:"))
        s.push(num)
        
    elif choice == 2:
        if s.isempty():
            print("Stack Underflow")
        else:
            num = s.pop()
            print("Popped data:", num)
            
    elif choice == 3:
        if s.isempty():
            print("Stack Underflow")
        else:
            num = s.peek()
            print("Top element in the stack:", num)
            
    elif choice == 4:
        if s.isempty():
            print("Stack Underflow")
        else:
            s.display()
            
    elif choice == 5:
        print("Quitting....")
        break
        
    else:
        print("Invalid choice!!!!! Enter the correct choice (1-5)! ")
        continue    
```
